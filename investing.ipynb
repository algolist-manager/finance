```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
import time
```

```python
start = time.time()
url = 'https://www.investing.com/stock-screener/Service/SearchStocks'

header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) '
                  'Chrome/79.0.3945.130 Safari/537.36',
    'X-Requested-With': 'XMLHttpRequest'
}

payload = {
    'country[]': '5',  ## US-country code
    'sector': '7,5,12,3,8,9,1,6,2,4,10,11', ## all sectors
    'industry': '81,56,59,41,68,67,88,51,72,47,12,8,50,2,71,9,69,45,46,13,94,102,95,58,100,101,87,31,6,38,79,30,77,'
                '28,5,60,18,26,44,35,53,48,49,55,78,7,86,10,1,34,3,11,62,16,24,20,54,33,83,29,76,37,90,85,82,22,14,'
                '17,19,43,89,96,57,84,93,27,74,97,4,73,36,42,98,65,70,40,99,39,92,75,66,63,21,25,64,61,32,91,52,23,'
                '15,80', ## all industries
    'equityType': 'ORD,DRC,Preferred,Unit,ClosedEnd,REIT,ELKS,OpenEnd,Right,ParticipationShare,CapitalSecurity,'
                  'PerpetualCapitalSecurity,GuaranteeCertificate,IGC,Warrant,SeniorNote,Debenture,ETF,ADR,ETC,ETN',
    'pn': 1,
    'order[col]': 'eq_market_cap',
    'order[dir]': 'd'
}

def run(pages):
    json_list = []
    for i in range(1, pages):
        try:
            payload['pn'] = i
            req_time = time.time()
            req = requests.post(url, data = payload, headers = header).json()
            req_end = time.time()
            print('소요시간 {}초'.format((req_end-req_time)/60))
            print('남은 페이지 수 : {}개'.format(pages - i))   
            json_list += req['hits']
        except Exception as e:
            print(e)
            
    df = pd.DataFrame(json_list)
    return df

pages = int(10914/50) + 1 ## 총 등록된 기업이 10914개이고 페이지 당 50개가 있으니 218페이지를 크롤링해야 합니다.
df = run(pages)

end = time.time()
print('{}분'.format((end-start)/60))
```

### 위의 코드를 바탕으로 10000개 종목의 일별 데이터를 가져올 수 있습니다. 그런데 시간이 너무 오래걸려서 31개 종목만 돌려보겠습니다. 31개를 돌리는데 37.96초가 나오므로 10000개라면 204분, 즉 3시간이 넘게 걸리네요... 1달 데이터만 가져오긴 하는 거지만 상당히 느리긴합니다....

```python
st = time.time()

tickers = list(result.index)
st_date = '01/12/2020'
end_date = '02/12/2020'

payload_stock = {
    'header': ticker + ' Historical Data', # 'VZ Histrical Data'와 같은 형식으로 ticker를 전달합니다.
    'st_date': st_date, # 01/12/2020 
    'end_date': end_date, # 02/12/2020
    'interval_sec': 'Daily', #일별 데이터입니다.
    'sort_col': 'date', #날짜별 정렬입니다.
    'sort_ord': 'DESC', #내림차순입니다.
    'action': 'historical_data'
}

daily_prices = []
for idx,ticker in enumerate(tickers):
    pair_id = result.loc[ticker]
    payload_stock['curr_id'] = pair_id
    payload_stock['header'] = ticker + ' Historical Data'
    print('{} 데이터 수집 중...'.format(ticker))
    req = requests.post(stock_url, data=payload_stock, headers=header).text
    soup = bs(req, 'lxml').find('table')
    daily_prices.append(str(soup))
    if idx==30:
        break

daily_price_stocks = " ".join(daily_prices)
daily_price_stocks_df = pd.read_html(daily_price_stocks)
end = time.time()
print(end-st)
```

### 아래의 url은 현재가를 받아오는 url입니다.

```python
url = 'https://advcharts.investing.com/advinion2016/advanced-charts/1/1/8/GetRecentHistory'
payload = {
    'strSymbol': 8127, 
    ## 이 symbol은 verizon의 pair_id입니다. 
    'iTop': '1500',
    'strPriceType': 'bid',
    'strFieldsMode': 'allFields',
    'strExtraData': 'lang_ID=1',
    'strTimeFrame': '5M' ## 5분 단위로 받아오겠습니다. 5분이 최소단위입니다.
}

header = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'
}
```

```python
iter_=0
while True:
    if iter_==0: ## 처음 시작할 떄는 00초의 정각이 될 때까지 기다리게 하겠습니다. 
        wait_time = 60 - datetime.now().second
        print('현재 시간 {0}, 대기시간 {1}초'.format(datetime.now().strftime('%H:%M:%S'), wait_time))
        time.sleep(wait_time)
        iter_+=1
    
    else:        
        req = requests.get(url, params=payload, headers=header).text
        latest = json.loads(req)['data'][-1]
        latest_df = pd.DataFrame([latest])
        now_minute = datetime.now().minute
        print(latest_df[['date','open','close', 'low', 'high', 'volume', 'volume_ac']])
        time.sleep(60) ## 1분 단위로 출력해보겠습니다.
```
